```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# 📊 Sample data (X = hours studied, y = exam score)
X = np.array([[1], [2], [3], [4], [5]])   # input feature
y = np.array([50, 55, 65, 70, 80])        # output

# 🧠 Create model
model = LogisticRegression()

# 📚 Train model
model.fit(X, y)

# 🔮 Predict
prediction = model.predict([[6]])

print("Predicted score for 6 hours study:", prediction[0])

# 📐 Model parameters
print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)
```

```python
import numpy as np
from sklearn.linear_model import SGDRegressor

# 📊 Sample data (X = hours studied, y = exam score)
X = np.array([[1], [2], [3], [4], [5]])   # input feature
y = np.array([50, 55, 65, 70, 80])        # output

# 🧠 Create model
shuvo = SGDRegressor()

# 📚 Train model
shuvo.fit(X, y)

# 🔮 Predict
prediction = shuvo.predict([[6]])

print("Predicted score for 6 hours study:", prediction[0])

# 📐 Model parameters
print("Slope (m):", shuvo.coef_[0])
print("Intercept (c):", shuvo.intercept_)
```

```python
import numpy as np
from sklearn.linear_model import SGDRegressor

# 📊 Sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([50, 55, 65, 70, 80])

# 🧠 Model
shuvo = SGDRegressor()

# 📚 Train
shuvo.fit(X, y)

# 📦 store predictions
predictions = []

# 🔮 Predict for 7, 8, 9 hours
for i in range(7, 10):
    pred = shuvo.predict([[i]])
    predictions.append(pred[0])
    print(f"Predicted score for {i} hours study: {pred[0]}")

# 📐 Model parameters
print("Slope (m):", shuvo.coef_[0])
print("Intercept (c):", shuvo.intercept_[0])
```
